```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Hảm hiển thị 2 ảnh khác nhau

def show_image(image1, image2, title1='1', title2='2'):
    f, ax = plt.subplots(nrows=1, ncols=2, figsize=(12, 12))
    ax[0].imshow(image1)
    ax[0].axis('off')
    ax[0].title.set_text(title1)
    ax[1].imshow(image2, cmap='gray')
    ax[1].axis('off')
    ax[1].title.set_text(title2)
    plt.show()
```

```python
# Đọc ảnh bằng openCV:
image = cv2.imread('dog.jpg')
# Chuyển ảnh từ màu BGR sang RGB:
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
# Hiển thị ảnh bằng matplotlib
plt.imshow(image)
plt.show()
```

```python
# Thực hiện resize ảnh: theo kích thước 384x384 (WxH)
image_resize = cv2.resize(image, (384, 384))
plt.imshow(image_resize)
```

```python
# Crop ảnh theo kích thước ngẫu nhiên
image_crop = image[200: 500, 500: 800]
plt.imshow(image_crop)

#cv2.imwrite('image_crop.jpg', cv2.cvtColor(image_crop, cv2.COLOR_RGB2BGR))
```

```python
# Xoay ảnh
image_rotate = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)
show_image(image, image_rotate, 'original', 'rotate 90')
```

```python
# Hàm chuyển đổi độ sáng của ảnh
# g(x,y)=a∗f(x,y)+b
def change_brightness(img, alpha, beta):
    img_new = np.asarray(alpha*img + beta, dtype=int)  
    img_new[img_new>255] = 255
    img_new[img_new<0] = 0
    return img_new

alpha = 1.0
beta = 100
image_brig = change_brightness(image, alpha, beta)
show_image(image, image_brig, 'original', 'after change brightness')
```

```python
# Hàm hiển thị ảnh nhị phân
def convert_to_binary(img, thresh=100):
    img_grayscale = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    thresh, img_binary = cv2.threshold(img_grayscale, thresh, maxval=255, type=cv2.THRESH_BINARY)
    return img_binary

image_binary = convert_to_binary(image)
show_image(image, image_binary, 'original', 'binary image')
```

```python
# Cân bằng độ sáng với openCV
equalized_image = cv2.equalizeHist(cv2.cvtColor(image, cv2.COLOR_RGB2GRAY))
show_image(image, equalized_image, 'original', 'image with equalize histogram')
```

```python
# Làm mờ ảnh với bộ lọc gauss
KERNEL_WIDTH = 15
KERNEL_HEIGHT = 15
SIGMA_X = 4
SIGMA_Y = 4
gauss_blur_image = cv2.GaussianBlur(image, ksize=(KERNEL_WIDTH, KERNEL_HEIGHT), sigmaX=SIGMA_X, sigmaY=SIGMA_Y)
show_image(image, gauss_blur_image, 'original', 'gauss blur image')
```

```python
KERNEL_SIZE = 15
median_blur_image = cv2.medianBlur(image, KERNEL_SIZE)
show_image(image, median_blur_image, 'original', 'median blur image')
```

```python
# Phát hiện cạnh với openCV
t_lower = 100
t_upper = 150
  
edge_image= cv2.Canny(image, t_lower, t_upper)
show_image(image, edge_image, 'original', 'edge image')

#cv2.imwrite('edge_image.jpg', edge_image)
```
